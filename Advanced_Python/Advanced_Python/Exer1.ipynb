```python
# 1. Given a matrix (list of lists), use list comprehension to return the transpose of the matrix.
# for example:
# For the input [[5,3,1] ,[2,3,6] ,[7,5,7] ,[2,9,6]]
# The output will be [[5, 2, 7, 2], [3, 3, 5, 9], [1, 6, 7, 6]]


def transpose(matrix: list) -> list:
    """ Transpose any matrix
    Args:
        matrix
    Returns:
        list[transposed list]
    """
    #returns transposed matrix using a for loop
    return [[row[index] for row in matrix] for index in range(len(matrix[0]))]


def main1():
    matrix = [[5, 3, 1], [2, 3, 6], [7, 5, 7], [2, 9, 6]]
    matrix1 = [[3, 4, 1], [2, 5, 6], [7, 9, 0], [1, 2, 5]]
    print(transpose(matrix))
    print(transpose(matrix1))


main1()

```

```python
# Use list comprehension to create a list that contains only the prime numbers
# in the range between 2 and n
# (hint: using the word all can help)


def find_all_prime_numbers_in_range(n: int) -> list:
    """ Finds all prime numbers in range n
    Args:
        n (int): [range]

    Returns:
        list: [list of prime numbers in range]
    """
    # move in given range and using module to determine if a number is prime if so adding it to the list
    return [num for num in range(2, n + 1)if all(num % index != 0 for index in range(2, int(num * 0.5) + 1))]


def main2():
    n = 5
    list_of_primes=find_all_prime_numbers_in_range(n)
    print(list_of_primes)


main2()
```

```python
# 3. Given a list of natural numbers,
# use list comprehension to create its sublist that contains the elements starting from the first index that have an odd number.
# for example:
# For the input [2,4,6,5,22,77,89]
# The output will be [22,77,89]


def get_all_numbers_from_first_odd_element_in_index(list: list) -> list:
    """ Getting all numbers after the first odd number encountered

    Args:
        list (list): [numbers]

    Returns:
        list: [all numbers after first odd ]
    """
    # using the next()function to find the index of the first odd element then using the : operator
    # getting all that comes after
    return list[next((index + 1 for index in range(len(list)) if list[index] % 2 != 0), len(list)):]
    


def main3():
    array = [2, 4, 6, 5, 22, 77, 89]
    first_odd_array=get_all_numbers_from_first_odd_element_in_index(array)
    print(first_odd_array)


main3()
```

```python


# 4. Create an expression generator that will return every time it is called
# the next number in the sequence of natural numbers from 1 to 100 which is the sum of squares of whole numbers (including 0).
# For example:
# in this list 25 appears since 25 is the sum of 4 squared (16) and 3 squared (9).
# 9 also appears since it is the sum of 3 squared (9) and 0 squared.
# For the output when we set sqr_gen to be the expression:
# for i in range(5):
# print(next(sqr_gen))
# we will receive: 1,2,4,5,8


def sqr_gen(max_n) -> int:
    """ Returns the next square root sum in the sequence
    Args:
        NONE
    Returns:
        next square root as int
    """

    return (n for n in range(1, max_n + 1) if any(index**2 + second_index**2 == n 
            for index in range(int(n**0.5) + 1) for second_index in range(int((n - index**2)**0.5) + 1)))


def main4():
    max_n = 100
    squares = sqr_gen(max_n)
    print("The squares that can be written as the sum of two squares are:", end=" ")
    for square in squares:
        print(square, end=" ")

main4()


```

```python
# 5. You are given a function func which performs some calculation on a natural number n
# and returns a natural number. The function comes from an external library
# and we have no information about it other than being able to use it.
# We need to create a list of all the outputs of the function from 1 to 1000.
# The problem is that for some of the outputs the function returns an error.
# How can we in a completely automatic way create such an output list
# for all valid outputs of the function, so that for valid outputs we get
# the output of the function and for errors we get -1?


def func(n: int) -> int:
    if not isinstance(n, int):
        return None
    if n % 2 == 0:
        return n
    else:
        raise Exception("Error: Odd number")


def create_output_list() -> list:
    """ Creates an output list for a given function with error handling
    Args:
        a given function
    Returns:
        output list with errors as -1
    """
    output_list = []
    for index in range(1, 1001):
        try:
            result = func(index)
            output_list.append(result)
        except Exception:
            output_list.append(-1)
    return output_list


def main5():
    print(create_output_list())


main5()
```

```python


# 6. Let's assume that some variable x is a number that represents age.
# Add code that will give an error if x entered is invalid
# (that is, either less than 15 or incomplete or greater than 80)
# Add code that will give an error if the entered x is not in the range between 15 and 80,
#  but if it is only incomplete (but in the range) then it is corrected to the whole value (without an error bounce)


def check_if_correct_age(age:int)->int:
    """ Check if the age is inside the limit if not raises exception

    Args:
        None:
    
    Returns:
        Prints the correction or the error
    """
    if age < 15 or age > 80:
        raise ValueError("Input value is not within the range of 15 to 80.")
    elif age.is_integer():
        print(age)
    else:
        age = int(age)
        print("the age is corrected to", age)


def main6():
    age=85.6
    check_if_correct_age(age)

main6()

```
