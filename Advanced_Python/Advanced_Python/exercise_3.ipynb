```python
#1.  Write a Python class to convert an integer to a roman numeral.
class IntegerToRoman:

    """A class for converting integers to Roman numerals.

    Attributes:
        roman_numerals (dict): A dictionary that maps integer values to Roman numeral symbols.

    Methods:
        __init__(self) -> None:
            Initializes a new IntegerToRoman object.

        int_to_roman(self, num: int) -> str:
            Converts an integer to a Roman numeral string.

    """

    def __init__(self) -> None:

        """Initializes a new IntegerToRoman object."""

        self.roman_numerals = {1000: 'M', 900: 'CM', 500: 'D', 400: 'CD', 100: 'C',
                          90: 'XC', 50: 'L', 40: 'XL', 10: 'X', 9: 'IX', 5: 'V',
                          4: 'IV', 1: 'I'}


    def int_to_roman(self:object, num: int) -> str:
        """Converts an integer to a Roman numeral string.

        Args:
            num (int): The integer to convert to a Roman numeral.

        Returns:
            str: The Roman numeral string representation of the input integer.
        """

        
        result = ''
        for value, numeral in self.roman_numerals.items():
            while num >= value:
                result += numeral
                num -= value
        return result
    
def main1():
    converted = IntegerToRoman()
    print(converted.int_to_roman(13564))

main1()
```

```python
#2. Write a Python class to convert a roman numeral to an integer.
class IntegerToRoman:

    """A class representing a converter for integers to Roman numerals.

    Attributes:
        roman_numerals (dict): A dictionary mapping Roman numerals to integers.

    Methods:
        __init__(self) -> None:
            Initializes a new IntegerToRoman object with the default Roman numeral mapping.

        roman_to_int(self, numeral: str) -> int:
            Converts the given Roman numeral to an integer.

    """
    def __init__(self) -> None:

        """Initializes a new IntegerToRoman object with the default Roman numeral mapping."""

        self.roman_numerals = {'M':1000,'CM':900,'D':500,'CD':400, 'C':100,
                          'XC':90,'L':50,'XL':40,'X':10,'IX':9,'V':5,
                          'IV':4,'I':1}
    
   
    def roman_to_int(self, numeral: str) -> int:

        """Converts the given Roman numeral to an integer.

        Args:
            numeral (str): The Roman numeral to be converted.

        Returns:
            int: The integer equivalent of the Roman numeral.
        """
        result = 0

        for index in range(len(numeral)):
            current_value = self.roman_numerals[numeral[index]]

            if index < len(numeral) - 1 and current_value < self.roman_numerals[numeral[index+1]]:
                result -= current_value
            else:
                result += current_value

        return result
    
def main2():
    converted = IntegerToRoman()
    print(converted.roman_to_int('MMMMMMMMMMMMMDLXIV'))
    
main2()
```

```python
#3. Write a Python class to find validity of a string of parentheses, '(', ')', '{', '}', '[' and ']. These brackets must be close in the correct order, for example "()" and "()[]{}" are valid but "[)", "({[)]" and "{{{" are invalid.

class check_validity_of_parentheses:
    """A class that checks validity of the parentheses in a string.

    Attributes:
        opening(list):a list with all opening parentheses
        closing(list):a list with all closing parentheses
        mapping(dic):a dictionary with all types 

    Methods:
        __init__(self) -> None:
            Initializes the three attributes with their respective assignment.

           
        is_valid(self, parentheses: str) -> bool:
            Checks if the parentheses are all closed properly .
        """
    def __init__(self) -> None:

        """
         Initializes the three attributes with their respective assignment.
        """
        self.opening = ['(', '{', '[']
        self.closing = [')', '}', ']']
        self.mapping = {')': '(', '}': '{', ']': '['}
        

    def is_valid(self, parentheses: str) -> bool:
        """
         Checks if the parentheses are all closed properly .

         Args:
            parentheses (str): The string which needs to be checked .

        Returns:
            bool: True or False.
        
        """
        stack = []
        for char in parentheses:
            if char in self.opening:
                stack.append(char)
            elif char in self.closing:
                if not stack or stack[-1] != self.mapping[char]:
                    return False
                stack.pop()
        
        return not stack

def main3():             
        parentheses = check_validity_of_parentheses()
        print(parentheses.is_valid('(())(()({[]())})' ))
        print(parentheses.is_valid(([({})])))
        
main3()       

```

```python
#4. Write a Python class to get all possible unique subsets from a set of distinct integers.
#Input : [4, 5, 6]
#Output : [[], [6], [5], [5, 6], [4], [4, 6], [4, 5], [4, 5, 6]]

class all_possible_subsets:

    """
    A class that returns all possible subsets for a given set.

    Attributes:
        subset(list):a subset that will contain all possible subsets

    Methods:
    __init__(self) -> None:
        Initializes the empty list.

        
    generate_subsets(self,nums:list) -> list:
        generates each subset 

    generate(self,current:list,nums:list) -> None:
        appends the current subset and calls the generate function recursively
    """


    def __init__(self)->None:
        """
          Initializes the empty list.
        """
        self.subset = []

    def generate_subsets(self,nums:list) -> list:

        """
         generates each subset 

         Args:
            nums (list): The list from which the  subsets are to ne generated.

        Returns:
            list: The list with all the subsets.
        
        """
        self.subsets = []
        self.generate([],nums)
        return self.subsets
    
    def generate(self,current:list,nums:list) -> None:
        """
        appends the current subset and calls the generate function recursively

        Args:
            nums(list): The list from which the  subsets are to be generated.
            current(list):The list that contains the current subset
        """
        
        self.subsets.append(current)
        for index in range(len(nums)):
            self.generate(current + [nums[index]], nums[index+1:])


def main4():
    sublist = all_possible_subsets()
    result = sublist.generate_subsets([4,5,6])
    print(result)

main4()
```

```python
# 5. Write a Python class to find a pair of elements (indices of the two numbers) from a given array whose sum equals a specific target number.
# Note: There will be one solution for each input and do not use the same element twice.
# Input: numbers= [10,20,10,40,50,60,70], target=50
# Output: 3, 4


class Find_two_elements_of_sum_n:

    """
    A class that returns the indices of the two first elements in array whose sum equals a given number .

    Methods:
    
        Find_the_corresponding_value(self,target_number:float,nums:list) -> tuple:
            Returns the two indices whose sum equals the given number
    """
    def Find_the_corresponding_value(self,target_number:float,nums:list) -> tuple:

        """
        Finds the pair of elements that equal to the specified value

        Args:
            target_number(float):The number for which we look for the sum
            nums(list):The list where we look for the elements
        Returns:
            a tuple with the two indices
        """
        try:
            for index in range(len(nums)):
                for second_index in range(index + 1, len(nums)):
                    if nums[index] == nums[second_index]:
                        continue
                    if nums[index] + nums[second_index] == target_number:
                        return index, second_index
            return None
        except TypeError:
            print("Error: The target_number and/or nums arguments are not of the correct data type.")
        except Exception as e:
            print(f"An error occurred: {e}")

def main5():
    Two_elements=Find_two_elements_of_sum_n()
    numbers= [10,20,10,40,50,60,70]
    target=30
    result= Two_elements.Find_the_corresponding_value(target,numbers)
    print(result)

main5()
```

```python
# 6. Write a Python class to find the three elements that sum to zero from a set of n real numbers.
# Input array : [-25, -10, -7, -3, 2, 4, 8, 10]
# Output : [[-10, 2, 8], [-7, -3, 10]]

class Find_triplets_sum_zero:

    """
    A class that finds every three numbers in a list whose sum equals 0 .

    Methods:
    find_three_elements(self,nums: list) -> list[list]:
        Creates a tuple for every three numbers whose sum equals 0 inside a list 

    """

    def find_three_elements(self,nums: list) -> list[list]:
        """
        A function to find three elements whose sum is 0

        Args:
            nums(list):The list of numbers where we look for the numbers

        Returns:
            A list pf list with each list containing three numbers whose sum is 0
            """
        try:
            nums.sort()
            length = len(nums)
            result = []
            for index in range(length):
                left_index = index+1
                right_index= length-1
                while left_index < right_index:
                    total = nums[index] + nums[left_index] + nums[right_index]
                    if total == 0:
                        result.append([nums[index], nums[left_index], nums[right_index]])
                        left_index += 1
                        right_index -= 1
                    elif total > 0:
                        right_index -= 1
                    else:
                        left_index += 1
            return result
        except TypeError:
            print("Error: The elements are not of the correct data type.")
        except Exception as e:
            print(f"An error occurred: {e}")
    
def main6():
    list_of_triplets=Find_triplets_sum_zero()
    Input_array = [-25, -10, -7, -3, 0, 2, 4, 0, 0, 8, 10,]
    result=list_of_triplets.find_three_elements(Input_array)
    print(result)

main6()
```

```python
# 8. Write a Python class to reverse a string word by word.
# Input string : 'hello .py'
# Expected Output : '.py hello'

class Reverse_string_word_by_word:
    """
    A class that Reverses a string word by  word rather than by letter

    Methods:
        reverse_words(self, string:str) -> str:
            reverses a given string by word 
    """
    def reverse_words(self, string:str) -> str:

        """
        A function to reverse a string word by word
        Args:
            string(str):The string to be reversed
        Returns: 
            reversed string
        """
        try:
            words = string.split()
            words.reverse()
            return ' '.join(words)
        except Exception as e:
            print(f"An error occurred: {e}")
    
def main8():
    reversed_string= Reverse_string_word_by_word()
    original_String = 'hello .py'
    result = reversed_string.reverse_words(original_String)
    print(result)

main8()
```

```python

# 10. Write a Python class named Rectangle constructed by a length and width and a method which will compute the area of a rectangle. 

class Rectangle:
    """
    A class that represents a rectangle and returns its area
        Attributes:
            length(float): the length of the rectangle
            width(float): the width of the rectangle 

        Methods:
        __init__(self,length:float,width:float) -> None:
            sets the length and width to the given value

         Compute_area_of_rectangle(self) -> float:
            computes the area of the rectangle
    """
    def __init__(self,length:float,width:float) -> None:
        """
        A function that sets the length and width of the rectangle

        Args:
            length(float):the length of the rectangle
            length(width):the width of the rectangle
        """
        self.length=length
        self.width=width

    def Compute_area_of_rectangle(self) -> float:

        """
        A function to compute the area of the rectangle
        Returns:
            the area of the rectangle
        """
        return self.length * self.width
    
def main10():
    rectangle=Rectangle(5,7)
    rectangle_area=rectangle.Compute_area_of_rectangle()
    print(rectangle_area)

main10()
```

```python
# 11. Write a Python class named Circle constructed by a radius and two methods which will compute the area and the perimeter of a circle. 

import math

class Circle:

    """
    A class that represents a circle and returns its area and perimeter
        Attributes:
           radius(float):the radius of the circle

        Methods:
        __init__(self,radius:float) -> None:
            sets the radius to the given value

            compute_area_of_circle(self) -> float:
            computes the area of the circle

            compute_perimeter_of_circle(self) -> float:
            computes the perimeter of the circle

    """
    def __init__(self,radius:float) -> None:
        """
        A function that sets the radius of the circle
        Args:
            radius(float):the circle's radius
        """
        self.radius=radius
    
    def compute_area_of_circle(self) -> float:
        """
        A function that computes the area of the circle
        Returns:
            the area of the circle
        """
        return self.radius**2 *math.pi
    
    def compute_perimeter_of_circle(self) -> float:
        """
        A function that computes the perimeter of the circle
        Returns:
            the perimeter of the circle
        """
        return 2*self.radius * math.pi
    
def main11():
    circle=Circle(12)
    circle_area=circle.compute_area_of_circle()
    print(circle_area)
    circle_perimeter=circle.compute_perimeter_of_circle()
    print(circle_perimeter)

main11()
```
