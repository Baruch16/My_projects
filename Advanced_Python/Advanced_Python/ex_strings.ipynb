```python
#1 Write a Python program that accepts a comma-separated sequence of words as input and prints the distinct words in sorted form (alphanumerically).
#INPUT: red, black, green
#OUTPUT: black, green,red
def Print_distinct_words(string:str) ->str:
    """
    Prints a sorted string of unique characters in string form

    Args:
        string(str):The string from which we take the uniques
    Returns:
        string of unique words
    """
    words_list = string.split(",") 
    unique_words = sorted(set(words_list)) 
    return ", ".join(unique_words) 
     

def main1():
    string=" red, black , green , red, black "
    distinct_words=Print_distinct_words(string)
    print(distinct_words)

main1()
```

```python
# 2 Write a Python program to print the following numbers up to 2 decimal places.
# INPUT: 
# x = 3.1415926
# y = 12.9999
# OUTPUT: 
# Formatted Number:       3.14
# Formatted Number:      13.00

def format_numbers(num: float, second_num: float) -> float:
    """
    Formats the given floating-point numbers to two decimal places and prints them.

    Args:
        num (float): The first floating-point number to format.
        second_num (float): The second floating-point number to format.
    """

    formatted_num = "{:>20.2f}".format(num)
    formatted_second_num = "{:>20.2f}".format(second_num)

    print(f"Formatted Number: {formatted_num}")
    print(f"Formatted Number: {formatted_second_num}")


def main2():
    num1 = 3.1415926
    num2 = 12.9999
    format_numbers(num1, num2)



main2()
```

```python
"""3 Write a Python program to display a number with a comma separator.
INPUT: 
x = 3000000
y = 30000000

OUTPUT: 
Formatted Number with comma separator: 3,000,000
Formatted Number with comma separator: 30,000,000
"""
def format_number_with_comma_separator(num: float) -> str:
    """
    Formats the given floating-point number with comma separators and returns it as a string.

    Args:
        num (float): The floating-point number to format.

    Returns:
        str: The formatted number as a string.
    """

    formatted_num = "{:,}".format(num)
    return f"Formatted Number with comma separator: {(formatted_num)}"


def main3():
    num = 3000000
    second_num = 30000000
    result = format_number_with_comma_separator(num)
    print(result)
    result = format_number_with_comma_separator(second_num)
    print(result)

main3()

```

```python
"""4 Write a Python program to print the following integers with zeros to the 
left of the specified width.
INPUT: 
x = 3
y = 123

OUTPUT: 
Formatted Number(left padding, width 2): 03
Formatted Number(left padding, width 6): 000123
"""
def format_number_with_left_padding(num: float) -> str:
    """
    Formats the given floating-point number with left padding, and returns it as a string.

    Args:
        num (float): The floating-point number to format.

    Returns:
        str: The formatted number as a string.
    """
    width = len(str(num)) * 2
    formatted_num = format(num, f"0{width}")
    return f"Formatted Number (left padding, width {width}): {formatted_num}" 


def main4():
    result = format_number_with_left_padding(3)
    print(result)
    result = format_number_with_left_padding(123)
    print(result)


main4()
```

```python
"""5 Write a Python program to swap commas and dots in a string.
INPUT: 
32.054,23

OUTPUT: 
32,054.23
"""
def replace_dots_and_commas(string: str) -> str:
    """
    Replaces all dots and commas in the input string, and returns the modified string.
    Dots are replaced with commas, and commas are replaced with dots.

    Args:
        string (str): The input string.

    Returns:
        str: The modified string with dots and commas replaced.
    """
    replacement = {'.': ',', ',': '.'}
    replacement_rule = str.maketrans(replacement)
    return string.translate(replacement_rule)

def main5():
        result=replace_dots_and_commas('32.054,23')
        print(result)

main5()
```

```python
"""6 Write a Python program to compute the sum of the digits in a given string.
INPUT: 
print(sum_digits_string("123abcd45"))
print(sum_digits_string("abcd1234"))

OUTPUT: 
15
10
"""

def sum_digits_string(string: str) -> int:
    """
    Sums up all the digits present in the input string and returns the total.

    Args:
        string (str): The input string to sum the digits of.

    Returns:
        int: The sum of all the digits present in the input string.
    """
    digits = [int(char) for char in string if char.isdigit()]
    return sum(digits)

def main6():
    result = sum_digits_string("123abcd45")
    print(result) 
    result =sum_digits_string("abcd1234")
    print(result)


main6()
```

```python
"""7 Write a Python program to find the maximum length of consecutive 0's in a given binary string.
INPUT: 
str1 = "111000010000110"
str2 = "111000111"

OUTPUT: 
Original string:111000010000110
Maximum length of consecutive 0's: 4

Original string:111000111
Maximum length of consecutive 0's: 3
"""

def find_the_maximum_consecutive_of_zero(string: str) -> str:
    """
    Finds the maximum length of consecutive 0's in the given string and returns the original string
    and the maximum length.

    Args:
        string (str): The input string to find the maximum length of consecutive 0's in.

    Returns:
        str: A string with the original input string and the maximum length of consecutive 0's.
    """
    max_len = 0
    current_len = 0
    for char in string:
        if char == '0':
            current_len += 1
            max_len = max(max_len, current_len)
        else:
            current_len = 0
    return f"Original string: {string}\nMaximum length of consecutive 0's: {max_len}\n"

def main7():
    string = "111000010000110"
    result=find_the_maximum_consecutive_of_zero(string)
    print(result)
    string='111000111'
    result = find_the_maximum_consecutive_of_zero(string)
    print(result)

main7()
```

```python
"""8 Write a Python program to find the smallest window that contains all characters in a given string.
INPUT: 
str1 = "asdaewsqgtwwsa"

OUTPUT: 
Original Strings: asdaewsqgtwwsa
Smallest window that contains all characters of the said string: daewsqgt
"""

from collections import defaultdict

def find_smallest_window(string: str) -> str:
    """
    Finds the smallest window in a given string which contains all the characters of another string.

    Args:
        string: The input string.

    Returns:
        The smallest window in the input string which contains all the characters of the string.

    """
    
    freq = defaultdict(int)
    length = len(string)
    unique_chars = len(set(string))
    start = 0
    min_len = length
    min_start = 0

    for index in range(length):
        freq[string[index]] += 1

        while len(freq) == unique_chars:
            if index - start + 1 < min_len:
                min_len = index - start + 1
                min_start = start
            freq[string[start]] -= 1
            if freq[string[start]] == 0:
                 del freq[string[start]]
            start += 1
    return "" if min_len == length else string[min_start:min_start+min_len]



def main8():
    str1 = "asdaewsqgtwwsa"
    str2='aacd'
    print("Original Strings:", str1)
    print("Smallest window that contains all characters of the said string:",find_smallest_window(str1))

main8()
```

```python
def find_smallest_window(string: str) -> str:
    """
    Finds the smallest window in the given string that contains all the distinct characters of that string.

    Args:
        string: The input string to be processed.

    Returns:
        The smallest window in the input string that contains all the distinct characters of that string.

    Example:
        >>> find_smallest_window("daewsqgt")
        "daewsqg"
    """
    end_pointer: int = len(string) - 1
    start_pointer: int = 0
    for char in string:
        if char in string[start_pointer + 1:]:
            start_pointer += 1
        else:
            break
    for char in reversed(string):
        if char in string[:end_pointer]:
            end_pointer -= 1
        else:
            break
    return string[start_pointer:end_pointer + 1]

def main8():
    str1 = "asdaewsqgtwwsa"
    str2='aacd'
    print("Original Strings:", str1)
    print("Smallest window that contains all characters of the said string:",find_smallest_window(str1))

main8()
```

```python
"""10 Write a Python program to convert a hexadecimal color code to a tuple of integers corresponding to its RGB components.
Use a list comprehension in combination with int() and list slice notation to get the RGB components from the hexadecimal string.
Use tuple() to convert the resulting list to a tuple.

INPUT: 
print(hex_to_rgb("FFA501"))
print(hex_to_rgb("FFFFFF"))
print(hex_to_rgb("000000"))
print(hex_to_rgb("FF0000"))
print(hex_to_rgb("000080"))
print(hex_to_rgb("C0C0C0"))

OUTPUT: 
(255, 165, 1)
(255, 255, 255)
(0, 0, 0)
(255, 0, 0)
(0, 0, 128)
(192, 192, 192)
"""

def hex_to_rgb(hex_code: str) -> tuple:
    """Converts a hexadecimal color code to an RGB color tuple.

    Args:
        hex_code: A string representing a hexadecimal color code.

    Returns:
        A tuple of three integers representing the red, green, and blue values of the color.
    """
    return tuple(
        int(hex_code[index : index + 2], 16)
        for index in range(0, len(hex_code), 2)
    )

    
def main10():
    print(hex_to_rgb("FFA501"))
    print(hex_to_rgb("FFFFFF"))
    print(hex_to_rgb("000000"))
    print(hex_to_rgb("FF0000"))
    print(hex_to_rgb("000080"))
    print(hex_to_rgb("C0C0C0"))

main10()
```

```python
"""11 Write a Python program to convert the values of RGB components to a hexadecimal color code.
Create a placeholder for a zero-padded hexadecimal value using '{:02X}' and copy it three times.
Use str.format() on the resulting string to replace the placeholders with the given values.

INPUT: 
print(rgb_to_hex(255, 165, 1))
print(rgb_to_hex(255, 255, 255))
print(rgb_to_hex(0, 0, 0))
print(rgb_to_hex(0, 0, 128))
print(rgb_to_hex(192, 192, 192))

OUTPUT: 
FFA501
FFFFFF
000000
000080
C0C0C0
"""
def rgb_to_hex(r: int, g: int, b: int) -> str:
    """Converts an RGB color tuple to a hexadecimal color code.

    Args:
        r: An integer representing the red value of the color.
        g: An integer representing the green value of the color.
        b: An integer representing the blue value of the color.

    Returns:
        A string representing the hexadecimal color code.
    """
    return '{:02X}{:02X}{:02X}'.format(r, g, b)


def main11():
    print(rgb_to_hex(255, 165, 1))
    print(rgb_to_hex(255, 255, 255))
    print(rgb_to_hex(0, 0, 0))
    print(rgb_to_hex(0, 0, 128))
    print(rgb_to_hex(192, 192, 192))

main11()
```
